```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans

df = pd.read_csv('Mall_Customers.csv')
df.head(5)
```

```python
#Se verifica que no existan valores nulos
print(df.isnull().sum())
print("==================")
df.shape

#Se eliminan las columnas CustomerID y Genero ya que no se consideran relevantes.
X = df.drop(columns=['CustomerID', 'Genre'])
X.head(5)
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

modelo = KMeans(n_clusters=3)
modelo.fit(X_std)
```

```python
df['clusters'] = modelo.labels_
df.head(10)
```

```python
sns.scatterplot(data=df, x='Age', y='Annual Income (k$)', hue='clusters')
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)', hue='clusters')
plt.show()
```
